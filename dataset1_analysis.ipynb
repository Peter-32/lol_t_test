```python
import pandas as pd
from scipy import stats
pd.set_option('display.max_columns', None)
df = pd.read_csv("dataset1/games.csv")
df['won'] = df['winner'].apply(lambda x: 1 if x == 1 else 0)
df = df[['won', 'firstBlood', 'firstTower', 'firstInhibitor', 'firstBaron', 'firstDragon', 'firstRiftHerald', 't1_towerKills', 't1_inhibitorKills', 't1_baronKills', 't1_dragonKills', 't1_riftHeraldKills', 't2_towerKills', 't2_inhibitorKills', 't2_baronKills', 't2_dragonKills', 't2_riftHeraldKills']].copy()
df
```

### Hypothesis: First blood has a stronger correlation with winning the game than securing first dragon
Null hypothesis: Securing first dragon has an equal or stronger correlation with winning the game than first blood.


```python
# Create groups based on firstDragon and firstBlood
dragon_wins = df[df['firstDragon'] == 1]['won']  # Check if Team 1 won when they got first dragon
blood_wins = df[df['firstBlood'] == 1]['won']  # Check if Team 1 won when they got first blood

# Perform two-sample T-test
t_stat, p_value = stats.ttest_ind(blood_wins, dragon_wins)

# For one-sided test, we check if first blood win rate > first dragon win rate
# Divide p-value by 2 for one-sided test and check if t-statistic is positive
one_sided_p_value = p_value / 2 if t_stat > 0 else 1 - (p_value / 2)

# Display the result
print(f"One-Sided T-statistic: {t_stat}, One-Sided P-value: {one_sided_p_value}")

if one_sided_p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

### Hypothesis: Securing the first Baron Nashor has a stronger correlation with winning than securing the first tower.
Null hypothesis: Securing first tower has an equal or stronger correlation with winning the game than taking first baron.


```python
# Create groups based on firstDragon and firstBlood
baron_wins = df[df['firstBaron'] == 1]['won']  # Check if Team 1 won when they got first dragon
tower_wins = df[df['firstTower'] == 1]['won']  # Check if Team 1 won when they got first blood

# Perform two-sample T-test
t_stat, p_value = stats.ttest_ind(baron_wins, tower_wins)

# For one-sided test, we check if first blood win rate > first dragon win rate
# Divide p-value by 2 for one-sided test and check if t-statistic is positive
one_sided_p_value = p_value / 2 if t_stat > 0 else 1 - (p_value / 2)

# Display the result
print(f"One-Sided T-statistic: {t_stat}, One-Sided P-value: {one_sided_p_value}")

if one_sided_p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

### Hypothesis: Taking 3+ towers has a stronger correlation with winning than securing the 2+ dragons.
Null hypothesis: Securing 2 or more dragons has an equal or stronger correlation with winning the game than taking 3 or more towers. 


```python
# Create groups for games where Team 1 secured 3 or more towers
tower_wins = df[(df['t1_towerKills'] >= 3)]['won']  # Wins with 3+ towers

# Create groups for games where Team 1 secured 2 or more dragons
dragon_wins = df[(df['t1_dragonKills'] >= 2)]['won']  # Wins with 2+ dragons

# Perform two-sample T-test (two-sided by default)
t_stat, p_value = stats.ttest_ind(tower_wins, dragon_wins)

# Display the result
print(f"Two-Sided T-statistic: {t_stat}, Two-Sided P-value: {p_value}")

one_sided_p_value = p_value / 2 if t_stat > 0 else 1 - (p_value / 2)


if one_sided_p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```
